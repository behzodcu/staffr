```python
from scipy import special
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
stats.norm.cdf(0)
# phi (norm-cdf) using erf
(1 + special.erf(0 / np.sqrt(2.0)))/2.0 
```

```python
x = np.linspace(-3, 3)
plt.plot(x, special.erf(x))
```

```python
x = np.linspace(-3, 3)
plt.plot(x, stats.norm.cdf(x))
plt.show()
plt.plot(x, (1 + special.erf(x / np.sqrt(2.0))) / 2.0 )
```
